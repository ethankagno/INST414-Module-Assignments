```python
import pandas as pd
```

```python
fangraphs = pd.read_csv("/Users/Jaron/Desktop/INST414 Files/Module Assignments/fangraphs_season_level.csv")
```

```python
fangraphs.head()
```

```python
data = fangraphs[fangraphs['Season'] == 2023]

data = data[data['IP'] >= 25]

data = data[['NameASCII', 'K_pct', 'BB_pct', 'H_per_9', 'xFIP', 'OSwing_pct', 'CSW_pct', 'HardHit_pct', 'Stuff_plus', 'Pitching_plus']]

data = data.set_index('NameASCII')

data.head()
```

```python
len(data)
```

```python
data_norm = data.divide(data.sum(axis=1), axis=0)
data_norm.head(10)
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

# Target Player 1: Logan Webb

```python
target_player1 = ('Logan Webb') 

target_player1
```

```python
target_player1_stats = data_norm.loc[target_player1]

target_player1_stats
```

```python
distances = scipy.spatial.distance.cdist(data_norm, [target_player1_stats], metric="euclidean")[:,0]
```

```python
query_distances = list(zip(data_norm.index, distances))

#query_distances.sort(key=lambda x: x[1])

#query_distances

comps = pd.DataFrame(query_distances, columns = ["Pitcher", "Distance"])

comps = comps.sort_values(by = ['Distance']).dropna()

comps.head(11)
```
